#2- Data Cleaning

```python
tips.isnull().sum()
```

```python
tips['tip%']= tips['tip']/tips['total_bill']*100
```

```python
tips.head()
```

```python
tips=tips.drop(['tips%'],axis=1)
```

```python
#3- Analyzing the relationship between variables
```

```python
correlation=tips.corr()
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(tips)
```
